```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar los datos
df = pd.read_csv('calidad_de_leche_vaca.csv')
```

```python
# Histograma de protena
plt.figure(figsize=(8,6))
plt.hist(df['Protena'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de contenido de protena')
plt.xlabel('Protena (%)')
plt.ylabel('Frecuencia')
plt.savefig('hist_proteina.png')

# Grfico de dispersin entre protena y grasa
plt.figure(figsize=(8,6))
plt.scatter(df['Protena'], df['Materia_grasa'], color='green')
plt.title('Relacin entre protena y materia grasa')
plt.xlabel('Protena (%)')
plt.ylabel('Materia Grasa (%)')
plt.savefig('scatter_proteina_grasa.png')

# Histograma de coliformes totales
plt.figure(figsize=(8,6))
plt.hist(df['Coliformes_totales'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribucin del conteo de coliformes totales')
plt.xlabel('Coliformes Totales (log UFC/mL)')
plt.ylabel('Frecuencia')
plt.savefig('hist_coliformes.png')
```
