```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar los datos
df = pd.read_csv('calidad_de_leche_vaca.csv')
```

```python
# Histograma de proteína
plt.figure(figsize=(8,6))
plt.hist(df['Proteína'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de contenido de proteína')
plt.xlabel('Proteína (%)')
plt.ylabel('Frecuencia')
plt.savefig('hist_proteina.png')

# Gráfico de dispersión entre proteína y grasa
plt.figure(figsize=(8,6))
plt.scatter(df['Proteína'], df['Materia_grasa'], color='green')
plt.title('Relación entre proteína y materia grasa')
plt.xlabel('Proteína (%)')
plt.ylabel('Materia Grasa (%)')
plt.savefig('scatter_proteina_grasa.png')

# Histograma de coliformes totales
plt.figure(figsize=(8,6))
plt.hist(df['Coliformes_totales'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribución del conteo de coliformes totales')
plt.xlabel('Coliformes Totales (log UFC/mL)')
plt.ylabel('Frecuencia')
plt.savefig('hist_coliformes.png')
```
